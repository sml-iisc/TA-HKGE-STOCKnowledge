```python
import requests
 
# Making a GET request

ticker = 'AAPL'
url = 'https://seekingalpha.com/symbol/' + ticker +'/news?filter=dividend_news&page=1'
url = 'https://seekingalpha.com/symbol/AAPL/news?filter=dividend_news&page=1'
r = requests.get(url)
 
# check status code for response received
# success code - 200
print(r)
 
# print content of request
print(r.content)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.content, 'html.parser')
print(soup.prettify())
```

```python
from   requests_html import AsyncHTMLSession
import functools

async def get_link(link):
    print('get   ', link)
    r = await asession.get(link)
    print('return', link)
    return r

asession = AsyncHTMLSession()

links = [
    'https://google.com',
    'https://yahoo.com',
    'https://python.org'
]

links = [ 
    functools.partial(get_link, link) for link in links
]

results = asession.run(*links)

print(results)
```

```python
from requests_html import HTMLSession
def func():
    
    session = HTMLSession()
    url = 'https://seekingalpha.com/symbol/AAPL/news?filter=dividend_news&page=1'
    r = session.get(url)
    r.html.render()
    return r

r = func()
print(r.html.raw_html)
```

```python
r.html.raw_html
soup = BeautifulSoup(r.html.raw_html, 'html.parser')
print(soup.prettify())
```
